## Area Between Two Curves

We learned that the integral

$$ \int_{a}^{b}f(x)dx $$

expresses the area enclosed by the function $f(x)$, with the x-axis between $x = a$ and $x = b$. The area between two curves is thus expressed as the integral

$$ \int_{a}^{b}(f(x) - g(x))dx $$

where $a$ and $b$ are the **points of intersection of the two curves** with $a < b$. The function $f(x)$ is referred to as the upper function and $g(x)$ as the lower function. 

## Example 1

Figure below illustrates this, assuming $f(x) = x$ and $g(x) = x^{2}$, with $a = 0$ and $b = 1$.

```python
from sympy import Symbol, Integral
x = Symbol('x')
f1 = x
f2 = x**2
Integral(f1-f2, (x, 0, 1)).doit().evalf()
```

## Example 2

Find the area between the curves  $y=x^{2}$  and  $y=x^{3}$ .

#### Solution

First we note that the curves intersect at the points  $(0,0)$  and  $(1,1)$ .

```python
from sympy import solve
x = Symbol('x')
expr = x**2 -x**3
solve(expr, dict=True)
```

Then we see that:

```python
from sympy.plotting import plot
from sympy import Symbol
x = Symbol('x')

p = plot(x**2, x**3, (x,-0.5,1.5), legend=True, show=False)
p[0].line_color = 'b'
p[1].line_color = 'r'

p.show()
```

In this interval. Hence the area is given by:

```python
from sympy import Symbol, Integral
x = Symbol('x')
f1 = x**2
f2 = x**3
Integral(f1-f2, (x, 0, 1)).doit().evalf()
```

## Example 3

Find the area between the curves  $y=0$  and  $y=3(x^{3}−x)$ .

```python
from sympy.plotting import plot
from sympy import Symbol
x = Symbol('x')

plot(3*(x**3 - x) , (x , -1.5 , 1.5))
```

#### Solution

When we graph the region, we see that the curves cross each other so that the top and bottom switch. Hence we split the integral into two integrals:

```python
from sympy import Symbol, Integral

x = Symbol('x')
f = 3*(x**3 - x)

area_left = abs(Integral(f, (x, -1, 0)).doit().evalf())
area_right = abs(Integral(f, (x, 0, 1)).doit().evalf())

print(area_left + area_right)
```
